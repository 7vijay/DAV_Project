# Training Baseline ML model

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=50, random_state=1)
#n_estimatiors=individual decision trees we want to train
#min_samples_split=individual decision tree shoudnt not split its node unless it has 50 samples
#random_state=ensures random numbers are generated in the same order if we run the same code code twice it is heplful
train = btc.iloc[:-200]#take all data expect the last 200 days
test = btc.iloc[-200:]# we take only the last 200 days into the test set

predictors = ["close", "volume", "open", "high", "low", "edit_count", "sentiment", "neg_sentiment"]  # to make sure we dont include tomorrow or target column which we created
model.fit(train[predictors], train["target"])
```

```python
from sklearn.metrics import precision_score  # to check how many time the model predicted correct

preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
precision_score(test["target"], preds)
```

# Evaluating error with backtesting

Above we tested only data for the last 200 days and figured the error
something weird might have happnd, so the model might not be oredicting well.
If we want to get predictions correct for as many rows as possible we can use backtesting.
Backtesting enables us to respect the order of the data but still getting predections for as many rows as possible

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="predictions")
    combined = pd.concat([test["target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=1095, step=150):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()  # copy is used to avoid warning
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
    # this will backtest our model against several years of data
```

Now we are going to upgrade our model from RandomForest.
RandomForest was good for just baseline model.
Now we use xgboost classifier.
it is a powerful Ml which can give us slightely more accurate predections

```python
from xgboost import XGBClassifier

model = XGBClassifier(random_state=1, learning_rate=.1, n_estimators=200)
#learningrate=defines if we want the model to protect against overfitting or not
#lower the learning rate, less likely the model is to overfit.
#xgboost uss something called boosting to connect the results of one model to the next
predictions = backtest(btc, model, predictors)
```

```python
predictions["predictions"].value_counts()
```

```python
precision_score(predictions["target"], predictions["predictions"])#compares actual values to predected values
```

```python
#adding some more predectiors
def compute_rolling(btc):
    horizons = [2,7,60,365]
    new_predictors = ["close", "sentiment", "neg_sentiment"]

    for horizon in horizons:
        rolling_averages = btc.rolling(horizon, min_periods=1).mean()

        ratio_column = f"close_ratio_{horizon}"
        btc[ratio_column] = btc["close"] / rolling_averages["close"]

        edit_column = f"edit_{horizon}"
        btc[edit_column] = rolling_averages["edit_count"]

        rolling = btc.rolling(horizon, closed='left', min_periods=1).mean()
        trend_column = f"trend_{horizon}"
        btc[trend_column] = rolling["target"]

        new_predictors+= [ratio_column, trend_column, edit_column]
    return btc, new_predictors
```

```python
btc, new_predictors = compute_rolling(btc.copy())
```
